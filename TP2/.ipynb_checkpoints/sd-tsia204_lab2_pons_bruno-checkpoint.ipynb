## Variable selection

## 2) Program the method of the forward variable selection

```python
# Define test of no effect in a general case. In our case the data is centered,
# so we could simplify some things but left the general case for further implementations if needed.

def test_no_effect (Y, X_k):
    
    n = X_k.size
    e1 = np.array([1,0])
    ones = np.full(n,1)
    
    # Fit a regular OLS
    reg = LinearRegression(fit_intercept=True).fit(X_k.reshape(-1,1), Y.reshape(-1,1))
    
    #Calculate the parameters for doing the statistic test
    G_n = 1/n * np.c_[ones,X_k].T @ np.c_[ones,X_k]
    s_n = np.sqrt(np.dot(np.dot(e1.T, np.linalg.inv(G_n)), e1))
    theta = np.array([reg.intercept_, reg.coef_[0]])
    sigma_n = 1/(n-2) * np.linalg.norm(Y - np.c_[ones,X_k] @ theta, ord=2)**2
    
    test = np.sqrt(n) * np.linalg.norm(theta, ord=1)/ (s_n * sigma_n)
    return test, theta

```

```python
# Defining forward variable selection without stop criterion

def fvs (X = X_strain, Y = Y_strain):
    # Number of variables p
    p = X_strain.shape[1]
    n = X_strain.shape[0]
    r = np.copy(Y)
    df = n - (p+1)
    
    #List for selected variables
    S = []
    
    #List for the p-values
    pval = []
    
    #Loop for doing fvs for every parameter
    for i in range(p):
        #List for containing every test value, reinitialized at every step
        T=[]
        thetas = []
        
        #Loop for doing the test for every parameter that has not been selected
        for k in range(p):
            if (k not in S):
                T.append(test_no_effect(r, X_strain[:,k])[0])
                thetas.append(test_no_effect(r, X_strain[:,k])[1])
            else:
                T.append(0)
                thetas.append(np.array([0,0]))
        
        #Selected index with the max test value
        idx = T.index(max(T))
        S.append(idx)
        r = r - np.c_[np.full(n,1),X_strain[:,idx]] @ thetas[idx]
        pval.append(2*(1-t.cdf(max(T),df)))
    
    S_array = np.asarray(S)
    p_val = np.asarray(pval)
    return S_array, p_val
```

## Ridge

## 5) Code your own ridge estimator using expression derived in class. Test it for a penalty parameter α spaced evenly on a log scale 10e-9 to 10e2.

```python

# Ridge estimator
def ridge (X_in , Y_in , lamda):
    X = np.copy(X_in)
    Y = np.copy(Y_in)
    n = X.shape[0]
    p = X.shape[1]
    return np.linalg.inv(X.T @ X +  n * lamda * np.eye(p)) @ X.T @ Y

```

```python

lamdas = np.logspace(-9, +2, num=50)
thetas = []
norms = []
normsRIDGE = []

for i in range(len(lamdas)):
    thetas.append(ridge(X_strain, Y_strain, lamdas[i]))
    norms.append(np.linalg.norm(thetas[i]))
    rid = Ridge(alpha = lamdas[i], fit_intercept=False)
    rid.fit(X_strain, Y_strain)
    normsRIDGE.append(np.linalg.norm(rid.coef_))
    
thetas_arr = np.asarray(thetas)
```

```python
fig, ax = plt.subplots()
for i in range(thetas_arr.shape[1]):
    plt.plot(lamdas, thetas_arr[:,i])
plt.xlabel("Lambda")
plt.ylabel("Values of the estimator coefficients")
ax.set_xscale("log")
plt.title("Coefficients in function of the parameter lambda")
plt.grid()
plt.show()
```

We can se in this plot that as lambda tends to large values the coefficients of the ridge estimator tends to zero, what is happening is that all the coefficients tends to 0. This is coherent with the theory as in ridge we are penalising for having large coefficients.

```python
fig, ax = plt.subplots()
plt.plot(lamdas, norms , color="r", label= "My ridge")
plt.plot(lamdas, normsRIDGE, label= "Sklearn ridge")
plt.xlabel("Lambda")
plt.ylabel("Norm of the coefficients")
ax.set_xscale("log")
ax.legend()
plt.title("Norm of the coefficients in function of the parameter lambda")
plt.grid()
plt.show()
```

We can se in this plot that as lambda tends to large values the norm of the ridge estimator tends to zero, what is happening is that the estimator tends to 0 and that is coherent with the theory. We can compare our result to the ridge estimator of sklearn and we observe they have the same behaviour but they are scaled differently, that can be explained on how we defined our minimization problem (the minimization problem in sklearn $||y - Xw||^2_2 + \lambda * ||w||^2_2$ , without a factor of n multiplying lambda as seen in class).

```python
mse_train = []
mse_test = []

for i in range(len(thetas)):
    Y_pred_train = X_strain @ thetas[i]
    Y_pred_test = X_stest @ thetas[i]
    mse_train.append(mean_squared_error(Y_strain,Y_pred_train))
    mse_test.append(mean_squared_error(Y_stest,Y_pred_test))
    
min_idx = mse_test.index(min(mse_test))
```

```python
fig, ax = plt.subplots()
plt.plot(lamdas, mse_train , color="b", label= "MSE for train data")
plt.plot(lamdas, mse_test , color="orange", label= "MSE for test data")
plt.plot(lamdas[min_idx], mse_test[min_idx], 'o', color='r' ,label= "Minimum" )
plt.xlabel("Lambda")
plt.ylabel("MSE")
ax.set_xscale("log")
plt.title("Mean square error in function of the parameter lambda")
ax.legend(loc="upper left")
plt.grid()
plt.show()
```

A question that arised while doing this part is why we plot both MSE for train data and test data, a possible answer for this question is because we want to check if the MSE for the test data follows the behaviour of the train data, this is certainly the case where it happens. We can also remark that when lambdas are large the mse tends to one. This is reasonable because the predicted values in this region are zero, so doing the mse is the same as calculating the variance for the Y values, and as the values are normalized the variance is equal to one.

```python
ridge = Ridge(alpha = lamdas[0], fit_intercept=False)
ridge.fit(X_strain, Y_strain)
R2ridge = ridge.score(X_stest, Y_stest)
```

## 7) Code your own version of the crossvalidation.

```python

# Define function for k-fold cross validation
def kfold_cv(model, X, y, k):
    # calculating the folds
    kf = KFold(n_splits=k)
    # list for saving the scores
    scores = []
    for train_index, test_index in kf.split(X):
        Xtrain, Xtest = X[train_index], X[test_index]
        ytrain, ytest = y[train_index], y[test_index]
        model.fit(Xtrain, ytrain)
        scores.append(model.score(Xtest, ytest))
    scores = np.asarray(scores)
    return scores.mean()

```

```python
R = np.logspace(-10, +3, num=30)
L = np.array([0,0.1,0.5,0.7,0.9,0.95,0.99])

scores = []

for i in range(len(R)):
    for k in range(len(L)):
        alpha = R[i] + L[k]
        l1 = L[k] / (2*R[i] + L[k])
        #eln = ElasticNet(alpha= R[i], l1_ratio=L[k], fit_intercept=False, max_iter=10000, tol=0.1)
        eln = ElasticNet(alpha= alpha, l1_ratio=l1, fit_intercept=False, max_iter=10000, tol=0.1)
        scores.append(kfold_cv(eln ,X_strain, Y_strain, 10 ))

```

```python
scores = np.asarray(scores)
print("Best R2 obtained with this method: ", max(scores))
```

So eventually the iteration are the same as by default and the convergence warnings arised as expected.

## 8)

```python
def Prediction_Interval(x, intercept, slope, sq_sigma ,X_mean, X_var ):
    y = intercept + slope*x
    interval = np.zeros(2)
    interval[0] = y - t.ppf(q, df)*np.sqrt(sq_sigma*(1 + 1/n + (x-X_mean)**2/X_var))
    interval[1] = y + t.ppf(q, df)*np.sqrt(sq_sigma*(1 + 1/n + (x-X_mean)**2/X_var))
    return interval
```

```python
p40 = np.asarray( df['V40'] ).reshape(-1,1)
y40 = np.asarray( Y ).reshape(-1,1)
```

```python

# Fit a regular OLS
reg = LinearRegression(fit_intercept=True).fit(p40, y40)
regline = np.asarray( reg.intercept_ )
regline = np.hstack([regline, np.asarray( reg.coef_)[0,0] ])
```

```python
fig, ax = plt.subplots()
plt.plot(p40, y40 , '.' , color="b", label= "Data")
plt.plot(p40, regline[0] + regline[1] * p40 , color="red", label= "Regression Line")
plt.xlabel("Values of parameter 40")
plt.ylabel("Y values")
plt.title("Linear regression plot for parameter 40")
ax.legend(loc="upper left")
plt.grid()
plt.show()
```

```python
bootstrap = []

for i in range(50):
    x1, y1 = resample(p40, y40, replace=True)
    # Fit a regular OLS
    reg1 = LinearRegression(fit_intercept=True).fit(x1, y1)
    regline1 = np.asarray( reg1.intercept_ )
    regline1 = np.hstack([regline1, np.asarray( reg1.coef_)[0,0] ])
    bootstrap.append(regline1)

bootstrap = np.asarray(bootstrap)
```

```python
n = len(p40)
alpha = 0.1
q= 1 - alpha/2
df= n - 2


# square sigma
sigma2 = 0
sigma2 = 1/(n-2) * np.sum((y40 - (regline[0] + regline[1]*p40))**2)


xmean = np.mean(p40)
xvar = np.sum((p40 - xmean)**2)

p40sorted = np.sort(p40.reshape(-1))

prediction_interval = []

for k in range(len(p40sorted)):
        auxP = np.zeros(2)
        auxP = Prediction_Interval(p40sorted[k], regline[0] , regline[1] , sigma2, xmean, xvar )
        prediction_interval.append(auxP)
prediction_interval = np.transpose(prediction_interval)
```

```python



fig, ax = plt.subplots(figsize=(10, 10))
plt.plot(p40, y40 , '.' , color="b", label= "Data")
plt.plot(p40, regline[0] + regline[1] * p40 , color="red" , label= "Regression Line")
plt.fill_between(p40sorted, prediction_interval[0], prediction_interval[1], color='red' , alpha=.1)
for i in range(len(bootstrap)):
    ax.plot(p40, bootstrap[i,0] + bootstrap[i,1] * p40, color="grey" , alpha=0.3)
    prediction_interval = []
    for k in range(len(p40sorted)):
            auxP = np.zeros(2)
            auxP = Prediction_Interval(p40sorted[k], bootstrap[i,0], bootstrap[i,1], sigma2, xmean, xvar )
            prediction_interval.append(auxP)
    prediction_interval = np.transpose(prediction_interval)
    ax.plot(p40sorted, prediction_interval[0], '--')
    ax.plot(p40sorted, prediction_interval[1], '--')
plt.xlabel("Values of parameter 40")
plt.ylabel("Y values")
plt.title("Linear regression plot for bootstrap of parameter 40")
ax.legend()
plt.grid()
plt.show()
```

## PCA

## 9) Compute the covariance matrix. Compute the singular value decomposition of the covariance matrix.

```python
#Calculate the covariance matrix
cov_matrix = np.cov(X_strain, rowvar=False)

# Computing the singular value decomposition of the covariance matrix
U, s, V = np.linalg.svd(cov_matrix, full_matrices=True)
```

```python
# Ploting the heatmap
fig, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(cov_matrix)
plt.title("Heatmap of the covariance matrix")
plt.show()
```

Above is the heatmap of the covariance matrix. When two covariates i and j have a strongly positive correlated, the value of the element will be close to 1.005 or in the heatmap to the white zones. If two covariates are less correlated the value of the element decreases and in the heatmap is seen as darker colors.

```python
# Plotting the singular values 
fig, ax = plt.subplots()
plt.plot(np.arange(2,len(s)+1), s[1:] , '.' , color="b", label= "Data")
plt.xlabel("Components")
plt.ylabel("Variance")
plt.title("Singular values between 2 to p.") #
ax.legend(loc="upper right")
plt.grid()
plt.show()
```

```python
pca = PCA()
pca.fit(X_strain)
variance = pca.explained_variance_
```

```python
# Plotting the amount of variance explained by the first k components
fig, ax = plt.subplots()
plt.plot(np.arange(2,len(s)+1), variance[1:] , '.' , color="r", label= "Data")
plt.xlabel("Components")
plt.ylabel("Variance")
plt.title("Amount of variance explained by first k components, k between 2..p .") #
ax.legend(loc="upper right")
plt.grid()
plt.show()
```

The aim for having both plots is to explain that, either using the singular values or using the sklearn PCA method to give the explained variance of each component give the same result. The statistical interpretation of singular values is in the form of variance in the data. Singular values produced by the SVD when squared are proportional to the amount of variance explained by a given singular vector.

```python

projected_data = X_strain @ U[:, :2]


fig, ax = plt.subplots()
plt.scatter(projected_data[:, 0], projected_data[:, 1], c=Y_strain)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.grid()
plt.show()
```

At a glance, we can see that it is very easy to diferenciate the data by its colors with reference to the principal component 1. As same color points represents close values in the data, we clearly can see that darker colors tend to be on the right side of the plot while lighter color points tend to be on the left.

This means that the principal component 1 is very useful at distinguishing the data, while the principal component 2 not so much. This was expected from the previous analysis. This plot can also help us identify possible outliers in our data.

```python
from sklearn.metrics import r2_score

best_score = -np.inf
best_k = None
p = X_strain[1].size

for k in range(1, p):
    projected_data = X_strain @ U[:, :k]
    reg = LinearRegression()
    reg.fit(projected_data, Y_strain)
    y_pred = X_stest @ (U[:, :k] @ reg.coef_.T)
    score = r2_score(Y_stest, y_pred)
    if score > best_score:
        best_score = score
        best_k = k

print(f'Best score: {best_score}, Best k: {best_k}')
```

```python
df_coef['R2 PCA before OLS'] = [best_score]
```

```python
df_coef
```

The best R2 we calculated is given by doing principal component analysis before applying ordinary least squares, we should stay with this approach in this case.
